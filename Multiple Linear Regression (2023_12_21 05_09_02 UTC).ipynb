```python
x = df.loc[:,['R&D Spend', 'Administration', 'Marketing Spend', 'City']].values
y = df.iloc[:,4:].values
```

```python
y
```

```python
#Convert categorical to numerical 
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
from sklearn.compose import ColumnTransformer #to use one hot encoding
label_encoder = LabelEncoder()
x[:,3] = label_encoder.fit_transform(x[:,3])
ohe_x = ColumnTransformer([('one_hot_encoder', OneHotEncoder(), [3])], remainder = 'passthrough') #passthrough tells that except the column '3', pass the other columns
x = np.array(ohe_x.fit_transform(x), dtype = int)
```

```python
x
```

```python
#dummy variable trap
x = x[:,1:]
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size = 0.3, random_state = 1)
from sklearn.linear_model import LinearRegression
regr = LinearRegression()
regr.fit(x_train, y_train)
y_train_pred = regr.predict(x_train)
y_test_pred = regr.predict(x_test)
from sklearn.metrics import mean_squared_error, r2_score
print(mean_squared_error(y_train, y_train_pred), mean_squared_error(y_test, y_test_pred))
print(r2_score(y_train,y_train_pred), r2_score(y_test, y_test_pred))
```
