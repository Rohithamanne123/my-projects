```python
from matplotlib import style
style.use('ggplot')
x=[1,2,3]
y=[4,5,6]
x1=[7,10,15]
y1=[4,9,6]
#plt.plot(x,y,color='green',linewidth=10,label='line 1')
plt.plot(x1,y1,color='magenta',linewidth=10,label='line 2')
plt.plot(x,y,'go-',linewidth=2,label='line 1')
#g is green, o is for round shape, - is for the line
plt.title('style')
plt.xlabel('x values')
plt.ylabel('y values')
plt.legend()
plt.grid(True,color='k')
plt.savefig('examprep.jpg')
```

```python
plt.bar(x,y,label='bar 1')
plt.bar(x1,y1,label='bar 2')
plt.grid(True,color='k')
plt.legend()
plt.title('bar graph')
plt.xlabel('x values')
```
