```python
#stack plot
days = [1,2,3,4,5]
sleeping = [7,8,6,11,7]
eating = [2,3,4,3,2]
working=[6,7,4,5,2]
playing = [8,5,4,3,2]

plt.plot([],[],color='m', label='sleeping', linewidth=2)
plt.plot([],[],color='c', label='eating', linewidth=2)
plt.plot([],[],color='r', label='working', linewidth=2)
plt.plot([],[],color='k', label='playing', linewidth=2)

plt.stackplot(days, sleeping, eating, working, playing, colors=['m','c','r','k'])
plt.title("Stack Graph")
plt.ylabel("Y-axis")
plt.xlabel('x-axis')
plt.legend()
plt.grid(True, color='k')
```

```python
#pie chart
days = [1,2,3,4]
activities = ['sleeping', 'eating', 'working', 'playing']
cols = ['m','c','r','b']
plt.pie(days,
 labels = activities,
 colors=cols,
 startangle=90,
 shadow = True,
 explode = (0.1,0.3,0.3,0.1),
 autopct = '%0.2f%%')
plt.title("pie chart")
plt.show()
```

```python
#multiple plots
import numpy as np
import matplotlib.pyplot as plt
def f(t):
 return np.exp(-t)*np.cos(2*np.pi*t)
t1 = np.arange(0.0,5.0,0.1)
t2 = np.arange(0.0,5.0,0.02)
plt.subplot(211) #2 - vertical 1 - Horizontal 1 - first graph
plt.plot(t1,f(t1),'bo',t2,f(t2))
plt.subplot(212) #2 - vertical 1 - Horizontal 1 - second graph
plt.plot(t2,np.cos(2*np.pi*t2))
plt.show()
```
