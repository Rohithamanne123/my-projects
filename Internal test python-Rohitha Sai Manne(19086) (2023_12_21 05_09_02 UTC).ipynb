```python
#[9] Write a Python program that creates a pie chart to represent the distribution of expenses in a household budget. Each category (e.g., rent, groceries, entertainment) should be a segment of the pie.
#9
import matplotlib.pyplot as plt
import numpy as np
expenses = ["rent", "groceries", "movies"]
cols = ['m','c','b']
amounts = np.random.randint(1000, 2000, size=3)
plt.pie(amounts, labels=expenses, autopct="%1.1f%%", colors=cols,counterclock=90,labeldistance=1.5,
 startangle=90,
 shadow = True,
 explode = (0,0.1,0)
 )
plt.title("Household budget")
plt.show()
```

```python
#[10] Write a Python program that generates a scatter plot with differently sized data points to 
#represent the population and GDP of different countries. Use population as the x-axis, 
#GDP as the y-axis, and adjust the size of the data points based on population.
populations = np.random.randint(10000, 100000,10)
gdp_per_capita = np.random.randint(100000, 500000,10)
gdp = populations * gdp_per_capita

plt.scatter(populations, gdp,s=populations/10,alpha=0.5,color="m",edgecolor="black")
plt.xlabel("Population (Millions)")
plt.ylabel("GDP (Billions)")
plt.title("Scatter Plot of Population vs. GDP")
plt.grid(True, color='k')
plt.show()
```
