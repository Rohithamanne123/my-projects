```python
x=df.iloc[:,0:1].values
y=df.iloc[:,1].values
print(x,y)
```

```python
#split the data
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=.25,random_state=20)
```

```python
x_train
```

```python
x_test
```

```python
y_train
```

```python
y_test
```

```python
#build the model
from sklearn.linear_model import LinearRegression
regr=LinearRegression()
regr.fit(x_train,y_train)
```

```python
y_train
```

```python
y_train_pred = regr.predict(x_train)
y_test_pred = regr.predict(x_test)
```

```python
y_train_pred
```

```python
y_test_pred
```

```python
from sklearn.metrics import mean_squared_error,r2_score
```

```python
#mean squared error
print('MSE train: %.3f,test: %.3f'%(mean_squared_error(y_train,y_train_pred),
                                    mean_squared_error(y_test,y_test_pred)))
```

```python
#r-squared
print('R^2 train: %8.3f,test:%.3f'%(r2_score(y_train,y_train_pred),r2_score(y_test,y_test_pred)))
```

```python
r2_score(y_test,y_test_pred)
```

```python
#visualize
plt.title('Linear Regression')
plt.xlabel('Exp')
plt.ylabel('Salary')
plt.scatter(x_train,y_train,color='red')
plt.plot(x_train,y_train_pred,color='blue',label='line1')
```
