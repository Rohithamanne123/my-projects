```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import random #to generate random numbers
import datetime #for date functions
import math #for math functions
%matplotlib inline
df=pd.read_excel('C:/Users/Rohitha Manne/Python Programming/Q1Dataset1.xlsx')
```

```python
correlation=df[['Quantity','Discount','Profit']].corr()
sns.heatmap(correlation,cbar=True,linewidth=5,annot=True,cmap='coolwarm')
```

```python
region_data=df.groupby('Region')[['Sales','Profit']].sum().reset_index()
sns.barplot(x='Region', y='Sales', data=region_data, label='Sales', color='blue')
sns.barplot(x='Region', y='Profit', data=region_data, label='Profit', color='green')
plt.xlabel('Region')
plt.ylabel('Amount')
plt.title('Sales and Profit by Region')
plt.legend()
```
