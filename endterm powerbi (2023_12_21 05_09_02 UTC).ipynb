```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import random #to generate random numbers
import datetime #for date functions
import math #for math functions
%matplotlib inline
df=pd.read_excel('C:/Users/Rohitha Manne/Dataset1_D1 (1).xlsx')
```

```python
# Data
states = ["Alabama", "Alaska", "Arizona", "Arkansas", "California", "Colorado", "Connecticut", "Delaware", "Florida", "Georgia"]
urban_percentage = [58, 65, 89, 56, 94, 86, 86, 83, 92, 74]
age_dependency = [64.6, 58.2, 67.9, 67.4, 59, 57.1, 60.9, 66.4, 67.7, 60.5]
poverty_percentage = [16.00, 10.30, 14.10, 16.10, 12.60, 9.80, 12.60, 12.60, 13.30, 14.30]
immigrants_percentage = [4, 8, 13, 5, 27, 10, 15, 10, 21, 10]
racial_distribution_texas = [40, 12, 39]
racial_labels = ["White alone (NH)", "Black or African American alone (NH)", "Hispanic or Latino (any race)"]
```

```python
plt.figure(figsize=(10, 6))
plt.bar(states, urban_percentage, color='purple')
plt.title('Urban Percentage (First 10 States)')
plt.xlabel('States')
plt.ylabel('Urban %')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.plot(states, age_dependency, marker='o', color='blue')
plt.title('Age Dependency Ratio (First 10 States)')
plt.xlabel('States')
plt.ylabel('Age Dependency Ratio')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(poverty_percentage, immigrants_percentage, color='orange')
plt.title('% of Population in Poverty vs. Immigrants % (First 10 States)')
plt.xlabel('% of Population in Poverty')
plt.ylabel('Immigrants %')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8, 8))
plt.pie(racial_distribution_texas, labels=racial_labels, autopct='%1.1f%%', startangle=90, colors=['lightgreen', 'lightpink', 'lightblue'])
plt.title('Racial Distribution for Texas')
plt.show()
```
